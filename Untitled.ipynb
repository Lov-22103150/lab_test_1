```python
import numpy as np
import matplotlib.pyplot as plt



def total_population(data :int):
    sum=0
    for count in data:
        sum+=count
    print(sum)    



def average_population_growth(data :int):
    sum=0
    for i in data:
        sum+=i
    print(sum/len(data))


def year_growth(data:int):
    sum=0
    for i in data:
        sum+=i
    for j in data:
        print(j/sum*100)    

            

def plotall(data:int,string :str):
    plt.plot(data,color="blue")
    plt.title(string)
    plt.xlabel("year")
    plt.ylabel("Population")


Lions=np.array([15,16,17,20,19,21,23,24,25,27])
Elephants=np.array([50,52,54,53,55,56,57,59,60,62])
Zebra=np.array([100,98,95,97,96,94,95,93,92,90])


print("total liones is : ")
total_population(Lions)
print("\n")
print("total elephents is :")
total_population(Elephants)
print("\n")
print("total Zebra is : ")
total_population(Zebra)


print("\n")
print("average yearly population growth liones is : ")
average_population_growth(Lions)
print("average yearly population growth Elephants is : ")
average_population_growth(Elephants)
print("average yearly population growth Zebra is : ")
average_population_growth(Zebra)


print("average year growth of lions :")
year_growth(Lions)
print("\n")
print("average year growth of Elephants :")
year_growth(Elephants)
print("\n")
print("average year growth of Zebra :")
year_growth(Zebra)
print("\n")


plotall(Lions,"Animals")
plotall(Elephants,"Animals")
plotall(Zebra,"Animals")

plt.bar(Lions,height=20.7)
plt.bar(Elephants,height=55.8)
plt.bar(Zebra,height=95.0)














```
